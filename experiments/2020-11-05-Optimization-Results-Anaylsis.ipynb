```python
from plotnine import *

import pyarrow.parquet as pq
import pyarrow as pa
import numpy as np
import pandas as pd
```

```python
df_1c1m = pq.read_table('../output/1c1m.parquet/').to_pandas()

# There appears to be some mistakes in 2 and 6 -- warrants some later investigation
#df_1c1m = df_1c1m.drop(df_1c1m[df_1c1m['input'] == 3].index)
#df_1c1m = df_1c1m.drop(df_1c1m[df_1c1m['input'] == 4].index)
```

```python
df_1c1m.groupby(['n_z', 'n_w']).mean().reset_index().drop('input', axis=1).sort_values(by='p_load_avg', ascending=False).head(10).values[0]
```

```python
df_1c1m.groupby('input').mean('p_load_avg').sort_values(by='p_load_avg')
```

```python
p = ggplot(aes(x='factor(n_z)', y='factor(n_w)', fill='p_load_avg'), df_1c1m.query('input == 6').groupby(['n_z', 'n_w']).mean().reset_index())
p + geom_tile() + scale_fill_cmap('turbo') + coord_equal() + theme(figure_size=(12, 12), axis_text_x=element_text(angle=90)) + ggtitle('1c1m') + labs(fill=r'$\bar{P}_{\mathrm{load}}$', x='nz', y='nw')
```

```python
df_1c1m.groupby(['n_z', 'n_w']).mean('p_load_avg').sort_values(by='p_load_avg', ascending=False)
```
